```python
# Detect outliers using the IQR method
outliers = {}
for column in numerical_columns:
    Q1 = sales_dataset[column].quantile(0.25)
    Q3 = sales_dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[column] = sales_dataset[(sales_dataset[column] < lower_bound) | (sales_dataset[column] > upper_bound)].shape[0]

print("Number of outliers in each column:\n", outliers)
```

# Standardization of Numerical Variables

### ●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.
### ●	Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
	
	Show before and after comparisons of the data distributions.


Standardization or Z-score normalization is a technique used in data preprocessing to rescale the features of a dataset so that they have the properties of a standard normal distribution with a mean of 0 and a standard deviation of 1.
Formula
The formula to calculate the Z-score for a data point 

 ## zi=σ−μ
 Where:
  
xi is the original data point.
μ is the mean of the data.

σ is the standard deviation of the data.

Steps in Standardization
Calculate the Mean: Determine the average value (mean) of the data.
Calculate the Standard Deviation: Measure the spread of the data points around the mean.
Apply the Z-Score Formula: For each data point, subtract the mean and then divide by the standard deviation to get the standardized value.
Interpretation
A Z-score tells you how many standard deviations a data point is from the mean.
Z = 0: The data point is exactly at the mean.
Z > 0: The data point is above the mean.
Z < 0: The data point is below the mean.
Z = 1: The data point is one standard deviation above the mean.
Z = -2: The data point is two standard deviations below the mean.

```python
df = pd.DataFrame(sales_dataset)

# Specified numerical columns
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 
                     'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']

```

```python
# Standardize each specified numerical column manually using the z-score formula
for column in numerical_columns:
    mean = df[column].mean()
    std = df[column].std()
    df[column] = (df[column] - mean) / std

print("Standardized Data:")
print(df)
```

```python
fig, axs = plt.subplots(len(numerical_columns), 2, figsize=(12, len(numerical_columns) * 4))

for i, column in enumerate(numerical_columns):
    sns.histplot(df[column].values, ax=axs[i, 0], kde=True)
    axs[i, 0].set_title(f'{column} - Before Standardization')
    
    mu = df[column].mean()
    sigma = df[column].std()
    standardized_column = (df[column] - mu) / sigma
    sns.histplot(standardized_column.values, ax=axs[i, 1], kde=True)
    axs[i, 1].set_title(f'{column} - After Standardization')

plt.tight_layout()
plt.show()
```

	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.

```python
df = pd.DataFrame(sales_dataset)

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Day', 'SKU', 'City', 'BU', 'Brand', 'Model'])

# Display a portion of the transformed dataset
print(df_encoded.head())
```

Converting categorical data into dummy variables (one-hot encoding) is essential for machine learning because:

1.Algorithm Compatibility: Most models require numerical input, and one-hot encoding transforms categorical data into a binary format that these models can process.

2.Avoiding Misinterpretation: One-hot encoding prevents the algorithm from mistakenly interpreting categorical data as ordinal or numeric, which could introduce biases.

3.Improving Performance: By clearly distinguishing categories, one-hot encoding helps models learn more effectively and improves performance.

4.Handling Multiple Categories: It allows each category to be represented separately, avoiding assumptions of numeric relationships.

5.Facilitating Feature Engineering: It helps in creating additional features or interactions that might benefit the model.
